```python
import tweepy
from textblob import TextBlob
import pandas as pd
import re
import numpy as np 
import matplotlib.pyplot as plt 
import csv
```

```python
data = pd.read_csv('pemindahanIbukota_Filtered.csv')
```

```python
#remove Username function
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
        
    return input_txt    
data.head()
```

```python
# remove twitter handles (@user)
data['tidy_tweet'] = np.vectorize(remove_pattern)(data['tweet'], "@[\w]*")
```

```python
# remove special characters, numbers, punctuations
data['tidy_tweet'] = data['tidy_tweet'].str.replace("[^a-zA-Z#]", " ")
```

```python
data['tidy_tweet'] = data['tidy_tweet'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
```

```python
#remove hashtag in front of words 
data['tidy_tweet'] = data['tidy_tweet'].str.replace("#([^\s]+)", " ")
```

```python
# Tokenization 
tokenized_data= data['tidy_tweet'].apply(lambda x: x.split())
tokenized_data.head()
```

```python
from nltk.stem.porter import *
# import Sastrawi package
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
tokenized_data = tokenized_data.apply(lambda x: [stemmer.stem(i) for i in x]) # stemming
tokenized_data.head()
```

```python
for i in range(len(tokenized_data)):
    tokenized_data[i] = ' '.join(tokenized_data[i])

data['tidy_tweet'] = tokenized_data
```

```python
import goslate
gs = goslate.Goslate()

dataMutable = data
dataMutable.shape
```

```python
# Remove URL
data['tidy_tweet'] = data['tidy_tweet'].str.replace("((www\.[^\s]+)|(https?://[^\s]+))", " ")
```

```python
import nltk
from nltk.corpus import stopwords
sw = stopwords.words('indonesian')
# Remove Stopwords Indonesia
data['tidy_tweet'] = data['tidy_tweet'].apply(lambda x: ' '.join([w for w in x.split() if w not in sw]))
```
